```python
h = ['#egalim','#egalimentation','#EtatsGenerauxdelAlimentation']
w = ["√©tats g√©n√©raux de l'alimentation"]
nb = 100
lg='fr'
```

```python
#Seach all the tweets in 1 week (max per search is 100, older access = 1 week)
search_results = twitter_api.search.tweets(q=h[0], count=nb,  result_type='recent')
statuses = search_results['statuses']
l_nw_st=100

while l_nw_st==100:
    mi=statuses[-1]['id']-1
    search_results = twitter_api.search.tweets(q=h[0], count=nb,  result_type='recent', max_id=mi)
    nw_st=search_results['statuses']
    l_nw_st=len(nw_st)
    statuses.extend(nw_st)
    l_st=len(statuses)
```

```python
conn = sqlite3.connect('tweets.db')
```

```python
c = conn.cursor()
# Create base
#------------------------------------------------------------
# Table: statuses
#------------------------------------------------------------
c.execute('''CREATE TABLE statuses(
        id                    Int NOT NULL ,
        coordinates           Varchar (250) ,
        created_at            Varchar (250) ,
        favorite_count        Int ,
        lang                  Varchar (25) ,
        place                 Varchar (2500) ,
        possibly_sensitive    Bool ,
        scopes                Varchar (2500) ,
        retweet_count         Int ,
        source                Varchar (2500) ,
        txt                   Varchar (250) ,
        withheld_in_countries Varchar (250) ,
        retweeted_status_id   Int ,
        in_reply_to_status_id Int ,
        quoted_status_id      Int ,
        user_id               Int ,
        PRIMARY KEY (id )
);''')


#------------------------------------------------------------
# Table: users
#------------------------------------------------------------
c.execute('''CREATE TABLE users(
        id                    Int NOT NULL ,
        screen_name           Varchar (250) ,
        contributors_enabled  Bool ,
        created_at            Varchar (100) ,
        default_profile       Bool ,
        default_profile_image Bool ,
        description           Varchar (2500) ,
        favourites_count      Int ,
        followers_count       Int ,
        friends_count         Int ,
        lang                  Varchar (25) ,
        listed_count          Int ,
        location              Varchar (250) ,
        name                  Varchar (25) ,
        protected             Bool ,
        statuses_count        Int ,
        PRIMARY KEY (id )
        );''')

#------------------------------------------------------------
# Table: hashtags
#------------------------------------------------------------

c.execute('''CREATE TABLE hashtags(
        htg Varchar (140) NOT NULL ,
        PRIMARY KEY (htg )
);''')



#------------------------------------------------------------
# Table: mentions
#------------------------------------------------------------

c.execute('''CREATE TABLE mentions(
        status_id Int NOT NULL ,
        user_id Int NOT NULL ,
        PRIMARY KEY (status_id ,user_id )
);''')


#------------------------------------------------------------
# Table: tags
#------------------------------------------------------------

c.execute('''CREATE TABLE tags(
        status_id Int NOT NULL ,
        htg    Varchar (140) NOT NULL ,
        PRIMARY KEY (status_id ,htg )
);''')
```
